```python
# Import library
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 2/kitten01.jpg') # Import gambar
plt.imshow(img) # Gambar menggunakan channel warna BGR
```

```python
# Konversi BGR ke RGB
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB) # Konversi BGR ke RGB
plt.imshow(img3)
```

```python
# Buat grayscale gambar
img_gray = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 2/kitten01.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
# Color map gray
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Color map magma
plt.imshow(img_gray, cmap = 'magma')
```

```python
# Resize gambar menjadi 512 x 1024
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

```python
# Flip gambar
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)

# Tampilkan plot dengan ukuran canvas yang lebih besar
fig = plt.figure(figsize = (10, 10)) # Ubah ukuran sesuai kebutuhan
ax = fig.add_subplot(111)
ax.imshow(img5)
```

```python
# Membuat Geometri 2D dari OpenCV
# Buat Black image dengan tipe data int16
black_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
plt.imshow(black_img)
```

```python
# Buat Persegi panjang

# Perhatikan koordinat pt1 dan pt2
cv.rectangle(black_img, pt1 = (384, 0), pt2 = (510, 150), color = (0, 255, 0), thickness = 10)
plt.imshow(black_img)
```

```python
# Buat Persegi panjang kedua

# Perhatikan koordinat pt1 dan pt2
cv.rectangle(black_img, pt1 = (200, 200), pt2 = (300, 300), color = (0, 0, 255), thickness = 15)
plt.imshow(black_img)
```

```python
# Buat lingkaran

cv.circle(black_img, center = (100, 100), radius = 50, color = (255, 0, 0), thickness = 8)
plt.imshow(black_img)
```

```python
# Buat garis sesuai koordinat pt1 dan pt2
cv.line(black_img, pt1 = (0, 0), pt2 = (512, 512), color = (255, 255, 255), thickness = 5)
plt.imshow(black_img)
```

```python
# Tambahkan text dengan font dan ukuran yang sudah ditentukan

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text = 'Hello', org = (10, 500), fontFace = font, fontScale = 4, color = (255, 255, 0), thickness = 2, lineType = cv.LINE_AA)
plt.imshow(black_img)
```

```python
# Black image dengan int32
black_img2 = np.zeros(shape = (512, 512, 3), dtype = np.int32)
plt.imshow(black_img2)
```

```python
# Buat array dengan Numpy
vertices = np.array([
    [100, 300],
    [200, 200],
    [400, 300],
    [200, 400]],
    dtype = np.int32)
vertices
```

```python
# Reshape array
pts = vertices.reshape((-1, 1, 2)) # nilai 2 menunjukkan bahwa tiap titik dibuat 3 channel yang memiliki R, G, B pts
pts
```

```python
# Menambahkan polyline
cv.polylines(black_img2, [pts], isClosed = True, color = (255, 0, 0), thickness = 5)
plt.imshow(black_img2)
```
