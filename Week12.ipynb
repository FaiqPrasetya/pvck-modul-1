```python
# Import Library
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Read Image
jokowi_ori = plt.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/jokowi.jpg')
kartini_ori = plt.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/kartini.jpg')
mycat_ori = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/mycat.jpg', 0) # 0 agar dibaca grayscale
mask_ori = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/mask.png', 0)
mjordan_ori = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/mjordan.jpg', 0)
solvay_ori = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/solvayconf.jpg')

# Pretrained feature
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/haarcascade_frontalface_alt.xml')
# Cascade classifier kucing yang diberikan melalui file melakukan hasil deteksi yang salah, jadi diubah menggunakan cv.data.haarcascade
cascade_kucing = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalcatface.xml')
```

## 2. Face Detection untuk image 

```python
jokowi = jokowi_ori.copy()

roi_wajah = cascade_wajah.detectMultiScale(jokowi_ori)

for(x, y, w, h) in roi_wajah:
  cv.rectangle(jokowi, (x, y), (x+w, y+h), (0, 255, 0), 3)

plt.figure(figsize = (10, 10))
f, axarr = plt.subplots(1, 2, figsize = (8, 8))
plt.subplot(121), plt.imshow(jokowi_ori)
plt.subplot(122), plt.imshow(jokowi)
plt.show()
```

### 6. Deteksi Wajah Gambar mjordan.jpg dan Tindih dengan smile.png

```python
mjordan_rgb = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/mjordan.jpg')
smiley = cv.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/smile.png')

# Convert BGR ke RGB karena output menggunakan plt
mjordan_rgb = cv.cvtColor(mjordan_rgb, cv.COLOR_BGR2RGB)
smiley = cv.cvtColor(smiley, cv.COLOR_BGR2RGB)

mjordan = mjordan_rgb.copy()

roi_wajah = cascade_wajah.detectMultiScale(mjordan_rgb)

for(x, y, w, h) in roi_wajah:
  smiley_resized = cv.resize(smiley, (w, h)) # resize smiley sesuai width dan height gambar mjordan
  mjordan[y:y+h, x:x+w] = smiley_resized # menimpa/menindih gambar mjordan dengan smiley_resized

plt.figure(figsize = (10, 10))
f, axarr = plt.subplots(1, 2, figsize = (5, 5))
plt.subplot(121), plt.imshow(mjordan_rgb)
plt.subplot(122), plt.imshow(mjordan)
plt.show()
```

## 7. Deteksi Wajah, Mata, dan Mata kanan

```python
people = plt.imread('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/people.jpg')

# Classififer Kedua Mata
cascade_mata = cv.CascadeClassifier('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/haarcascade_eye.xml')

# Classifier Mata Kanan
cascade_mata_kanan = cv.CascadeClassifier('/content/drive/MyDrive/Kuliah/Semester 6/PVCK/Pertemuan 12/Face Detection/haarcascade_righteye_2splits.xml')

# Load classifier pre-trained 
roi_wajah = cascade_wajah.detectMultiScale(people)
roi_mata = cascade_mata.detectMultiScale(people)
roi_mata_kanan = cascade_mata_kanan.detectMultiScale(people)

# Deteksi Wajah
for(x, y, w, h) in roi_wajah:
  cv.rectangle(people, (x, y), (x+w, y+h), (0, 0, 0), 1)

# Deteksi Kedua Mata
for(x, y, w, h) in roi_mata:
  cv.rectangle(people, (x, y), (x+w, y+h), (255, 255, 255), 1)

# Deteksi Mata Kanan
for(x, y, w, h) in roi_mata_kanan:
  cv.rectangle(people, (x, y), (x+w, y+h), (255, 0, 0), 2)

plt.imshow(people)
```
